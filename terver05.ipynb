1\. Когда используется критерий Стьюдента, а когда Z --критерий?

Z-критерий используется тогда, когда известна σ (среднеквадратичое
отклонение) генеральной совокупности и генеральная совокупность имеет
нормальное распределение.

t-критерий (криитерий Стьюдента) применяется тогда, когда совокупность
имеет нормальное распределение, но σ генеральной совокупности не
известна, а дисперсии равны.

При больших размерах выборки Z-тест и t-тест дают схожие значения
p-value.

3\. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки
печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес
каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194,
190. Известно, что их веса распределены нормально. Верно ли утверждение
продавца, если учитывать, что доверительная вероятность равна 99%?
(Провести двусторонний тест.)

Поскольку нам не известно среднеквадратичное отклонение генеральной
совокупности σ, то проверку гипотез будем проводить с помощью критерия
Стьюдента (t-тест). Критерий α двусторонний рассчитывается как:

$\alpha = 1 - \left( \frac{99}{100} \right) = 0,01;\frac{\alpha}{2} = 0,005$.

Для выборки рассчитаем среднее, стандартное несмещенное отклонение,
количество степеней свободы и значение критерия Стьюдента:

${x} = \frac{\sum_{}^{}x}{n};\sum_{}^{}x = 1985,0000;n = 10;$

${x} = 198,5000; \sigma_{Н} = \sqrt{\frac{1}{n - 1} \cdot \sum_{i = 1}^{n}\left( x_{i} - \overline{x} \right)^{2}};\sum_{i = 1}^{n}\left( x_{i} - \overline{x} \right)^{2} \approx 187,4198; \Rightarrow \sigma_{Н} = \sqrt{\frac{187,4198}{9}} \approx 4,5634;$

$k = n - 1 = 10 - 1 = 9;$

$t_{Н} = \frac{\overline{X} - \mu}{\frac{\sigma_{Н}}{\sqrt{n}}};\overline{X} = 198,5;\mu = 200;\sigma_{Н} = 4,5634;n = 10; \Rightarrow t_{Н} = - 1,0394;$

Табличное значение $t_{Т} \approx 3,6897$при $\frac{\alpha}{2} = 0,005$,
таким образом $t_{Н} = 1,0394 < t_{Т} = 3,6897$следовательно,
принимается гипотеза $H_{0}$, статистически значимые различия не
обнаружены.

**Ответ: утверждение о том, что средняя масса пачки печенья составляет
200 г соответствует истине при α = 0,01.**


4. Есть ли статистически значимые различия в росте дочерей?
Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
import numpy as np
from scipy import stats

x1 = np.array([172, 177, 158, 170, 178, 175, 164, 160, 169])
x2 = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])
if stats.ttest_rel(x1, x2).pvalue > 0.05:
    print ("Статестически значемые различия не обнаружены")
else:
    print ("Обнаружены статистически значемые различия")
```
